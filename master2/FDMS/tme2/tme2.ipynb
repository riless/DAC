## Classe Data()
Cette classe est créée de sorte a pouvoir télécharger automatiquement des données depuis http://mldata.org grace à la classe <i>sklearn.datasets.mldata.fetch_mldata</i> et à les binarisé (garder seulement deux classes) pour les besoins de ce TME.

Des fonctions de génération de données et de visualisation (pour les données avec deux attributs) sont également implémentées.

```python
%matplotlib inline

from sklearn.datasets.mldata import fetch_mldata
import sys
import tempfile
import numpy as np
import random
import matplotlib.pyplot as plt

class Data():
    def __init__(self, dataname="generate", keep_one_to = 1, nbex=500, centerx=1,centery=1,sigma=0.1,epsilon=1):
        if( dataname=='generate' ):    
            self.x, self.y = self.generate(nbex=nbex, centerx=centerx,centery=centery,sigma=sigma,epsilon=epsilon);
        else:
            dh = tempfile.mkdtemp()
            
            try:
                d = fetch_mldata(dataname=dataname, data_home=dh)
            except:
                print "/!\ erreur lors du chargement de la dataset <%s> depuis mldata.org" % dataname
                sys.exit(1)
                
            y = d.target
            y[y != keep_one_to] = -1 
            self.x = d.data
            self.y = y
    
    def get(self):
        return self.x, self.y
    
    def shape(self):
        return self.x.shape, self.y.shape

    def draw(self):
        dim = self.shape()[0][1]
        if ( dim != 2 ):
            print "/!\ Impossible d'afficher %d dimensions\nSeules les deux premières seront affichées" % dim
        
        plt.scatter(self.x[self.y<0,0],self.x[self.y<0,1], c='red',marker='o')
        plt.scatter(self.x[self.y>0,0],self.x[self.y>0,1], c='green',marker='o')
            
    
    def generate(self, nbex=100, centerx=1,centery=1,sigma=0.75,epsilon=0.5):
         xpos=np.random.multivariate_normal([centerx,centerx],np.diag([sigma,sigma]),nbex/2)
         xneg=np.random.multivariate_normal([-centerx,-centerx],np.diag([sigma,sigma]),nbex/2)
         data=np.vstack((xpos,xneg))
         y=np.hstack((np.ones(nbex/2),-np.ones(nbex/2)))

         data[:,0]+=np.random.normal(0,epsilon,nbex)
         data[:,1]+=np.random.normal(0,epsilon,nbex)

         idx = np.random.shuffle((range(y.size)))
         data=data[idx,:]
         y=y[idx]
         return data[0],y[0]
```

## Classe LR()

La classe LR ( où Linear Regression ) est un classifieur linéaire avec descente de gradient dans laquelle est implémenté la Régularisation L1.

Elle dispose d'une méthode <i>fit()</i>permettant d'entrainer le modèle. et des fonctions <i>predict()</i>, <i>sparcity()</i>, <i>error()</i>, <i>accuracy()</i> qui permettent respectivement de faire de la prédiction, de renvoyer le nombre de données nules dans le modèle, de renvoyer l'erreur au carré sur les données d'apprentissage, et enfin le taux de bonne classification sur les données d'entrées.

```python
class LR():
    
    def __init__(self, iter=100, e=0.001, lmbda=0.2, debug=False, debug_interval=20):
        self.iter = iter
        self.e = e
        self.lmbda = lmbda
        self.seuil = self.e * 0.00001; # seuil pour arrêter la descente de gradient 
        self.debug_interval = debug_interval
        
        self.debug = debug # activer ou non les info de debug
         
        self.error_history = []
        self.theta_history = []
    
    def _print_debug(self, it=0):
        if (self.debug and it%self.debug_interval==0):
            print "#%d | Taux bonne classification %.2f | erreur²: %.2f" % (it, self.accuracy(x_data,y_data), self.error(x_data,y_data) )

    def draw(self):
        x_plot = np.arange(-5, 5, 0.1);
        for idx, theta in enumerate( self.theta_history ):
            ls = '--'
            lw = 1
            if ( idx == len(self.theta_history)-1 ):
                y_plot = -1 * (theta[1]*x_plot + theta[0]) # -1 pour avoir la perpondiculaire
                plt.plot(x_plot, y_plot,ls='-', lw=2 , label="Theta Final (it=%d)"%idx )
            elif (idx % 20 == 0):
                y_plot = -1 * (theta[1]*x_plot + theta[0]) # -1 pour avoir la perpondiculaire
                plt.plot(x_plot, y_plot,ls='--', lw=1 , label="Iteration "+str(idx) )
        plt.legend()
        
    def fit(self, x_data, y_data):

        it = 0
        converged = False
        m, n = x_data.shape
        self.theta = np.random.rand(n)

        while it < self.iter and not converged:
            
            self.error_history.append( self.error(x_data, y_data) ) # erreur avec theta actuel
            self.theta_history.append( self.theta )
            arr = np.random.permutation(m)
            for idx in arr:
                x = x_data[idx]
                y = y_data[idx]

                gr = -2 * ( y - self.theta * x ) * x # Gradient
                gr += self.lmbda * np.sign(self.theta) # + L1
                
                new_theta = self.theta - self.e * gr # update theta
                self.theta = np.where( (new_theta * self.theta < 0), 0, new_theta ) # Gradient Clipping
                
            self._print_debug(it) # info de debug
  
            if ( self.error_history[it] - self.error(x_data, y_data) ) < self.seuil: # arrêter si n'avance pas assez vite
                converged = True
                
                # last iteration
                self.error_history.append( self.error(x_data, y_data) )
                self.theta_history.append( self.theta )
                self._print_debug(it)
                
            it+=1
        return self.theta

    
    def sparcity(self):
        return len(self.theta) - np.sum( np.where( self.theta != 0, 1, 0) ) # nombre de params != 0

    
    def predict(self, x):
        return np.where( x.dot(self.theta) < 0, -1, 1 )
    
    
    def accuracy(self, x, y): 
        return np.sum( y==self.predict(x) )/float( len(x) ) # nombre d'elements bien classifiés
    
    
    def error(self, x, y):
        return np.sum( ( y - x.dot(self.theta) ) ** 2 ) / len(x) + self.lmbda * np.sum( np.absolute( self.theta )) 
```

## Algorithme de descente du gradient

L'algorithme de descente du gradient consiste a chercher un &theta; de sorte à ce que la fonction de décision f&theta;(x) colle le plus souvent aux vrais labels y.

Dans notre cas la fonction de décision f&theta; est juste un produit scalaire entre x ( les attributs de chaque exemple d'apprentissage ) et &theta;  

Pour trouver le paramètres &theta; optimal, il nous faut un critère d'apprentissage (une donnée qui nous dit si on s'est plus ou moins trompé dans la classification).

## La fonction coût
La fonction coût que nous allons utiliser est une fonction des moinre carrée avec un paramètre $\lambda C(\theta )$ ou $C(\theta) $ est le nombre de feature utilisé et $\lambda $ un coefficient choisis manuellement qui vas nous permettre de faire un compromis entre la qualité du modèle et le nombre de features utilisées.

Cette fonction depend aussi de la fonction <i>predict</i> qui représente ici la fonction de décision $f_\theta(x)$ et qui est juste un produit scalaire entre $\theta$ et les exemples d'apprentissage x
